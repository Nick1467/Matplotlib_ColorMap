````python
import numpy as np
import matplotlib.pyplot as plt
import os

# üîß ‰Ω†ÂèØ‰ª•Âú®ÈÄôË£°ÊâãÂãï‰øÆÊîπ colormap ÂàóË°®ÔºàÊúÄÂ§öÂª∫Ë≠∞ 12 ÂÄãÔºâ
colormaps_to_show = [
    "viridis", "plasma", "inferno", "magma", "cividis",
    "Greys", "Blues", "RdBu", "seismic", "coolwarm",
    "tab10", "nipy_spectral"
]

def load_data(filepath):
    ext = os.path.splitext(filepath)[1]
    if ext == ".npy":
        return np.load(filepath)
    elif ext == ".csv":
        return np.loadtxt(filepath, delimiter=",")
    else:
        raise ValueError("Unsupported file format. Please use .npy or .csv")

def plot_colormaps(data, cmap_list):
    n = len(cmap_list)
    cols = 4  # ÂπæÊ¨ÑÂúñÔºàË¶ñÈúÄË¶ÅË™øÊï¥Ôºâ
    rows = int(np.ceil(n / cols))

    fig, axes = plt.subplots(rows, cols, figsize=(4 * cols, 3 * rows))
    axes = axes.flatten()

    for i, cmap in enumerate(cmap_list):
        ax = axes[i]
        im = ax.imshow(data, cmap=cmap)
        ax.set_title(cmap, fontsize=10)
        ax.axis("off")
        fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    # ÊääÂâ©È§òÁ©∫ÁôΩÂúñÂèñÊ∂à
    for j in range(i + 1, len(axes)):
        axes[j].axis("off")

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    import sys

    print("Ë´ãËº∏ÂÖ• 2D array Ê™îÊ°àË∑ØÂæëÔºàÊîØÊè¥ .npy Êàñ .csvÔºâÔºö")
    path = input("> ").strip()

    try:
        data = load_data(path)
        if data.ndim != 2:
            raise ValueError("Ëº∏ÂÖ•ÁöÑË≥áÊñô‰∏çÊòØ 2D array")
    except Exception as e:
        print(f"‚ùå Ë≥áÊñôËÆÄÂèñÈåØË™§Ôºö{e}")
        sys.exit(1)

    print(f"‚úÖ Ë≥áÊñôËºâÂÖ•ÊàêÂäüÔºåË≥áÊñô shape = {data.shape}")
    plot_colormaps(data, colormaps_to_show)
````

````python
import numpy as np
import matplotlib.pyplot as plt
import os

# È†êË®≠Â±ïÁ§∫ÁöÑ colormaps
colormaps_to_show = [
    "viridis", "plasma", "inferno", "magma", "cividis",
    "Greys", "Blues", "RdBu", "seismic", "coolwarm",
    "tab10", "nipy_spectral"
]

def load_data(filepath):
    ext = os.path.splitext(filepath)[1]
    if ext == ".npy":
        return np.load(filepath)
    elif ext == ".csv":
        return np.loadtxt(filepath, delimiter=",")
    else:
        raise ValueError("Unsupported file format. Please use .npy or .csv")

def generate_sample_data():
    x = np.linspace(-3, 3, 150)
    y = np.linspace(-3, 3, 150)
    X, Y = np.meshgrid(x, y)
    Z = np.sin(X**2 + Y**2)
    return Z

def plot_colormaps(data, cmap_list):
    n = len(cmap_list)
    cols = 4
    rows = int(np.ceil(n / cols))

    fig, axes = plt.subplots(rows, cols, figsize=(4.5 * cols, 4 * rows))
    axes = axes.flatten()

    for i, cmap in enumerate(cmap_list):
        ax = axes[i]
        im = ax.imshow(data, cmap=cmap)
        ax.set_title(cmap, fontsize=25, fontweight='bold')  # ‚¨Ö Â≠óÈ´îËÆäÂ§ßÊõ¥Ê∏ÖÊô∞
        ax.axis("off")
        fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    for j in range(i + 1, len(axes)):
        axes[j].axis("off")

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    print("ÈÅ∏ÊìáË≥áÊñô‰æÜÊ∫êÔºö")
    print("1. ÂåØÂÖ• 2D Èô£ÂàóË≥áÊñôÔºà.npy Êàñ .csvÔºâ")
    print("2. ‰ΩøÁî®ÂÖßÂª∫ÁØÑ‰æãË≥áÊñôÔºàsin(x¬≤ + y¬≤))Ôºâ")
    choice = input("> Ëº∏ÂÖ• 1 (ÂåØÂÖ•) Êàñ 2 (ÂëºÂè´ËÆäÊï∏)Ôºö ").strip()

    if choice == "1":
        path = input("Ë´ãËº∏ÂÖ•Ê™îÊ°àË∑ØÂæëÔºà.npy Êàñ .csvÔºâÔºö ").strip()
        try:
            data = load_data(path)
            if data.ndim != 2:
                raise ValueError("‚ö†Ô∏è Ëº∏ÂÖ•Ë≥áÊñô‰∏çÊòØ 2D Èô£Âàó")
        except Exception as e:
            print(f"‚ùå ÈåØË™§Ôºö{e}")
            exit(1)
    else:
        print("‚úÖ ‰ΩøÁî®ÂÖßÂª∫ÁØÑ‰æãË≥áÊñô")
        data = generate_sample_data()

    print(f"üìê Ë≥áÊñôÂ§ßÂ∞èÔºö{data.shape}")
    plot_colormaps(data, colormaps_to_show)
````
